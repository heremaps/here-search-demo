# App automation

```python
from here_search_demo.widgets.app import OneBoxMap
app = OneBoxMap(
    route_post = True
)
app.show()
app.run()
```

```python
import asyncio

box = app.query_box_w                               # Text box inside OneBoxMap
response_items = app.map_w.state.items_data_by_rank # The search response items
route_ctl = app.map_w.route                         # The route controller

# Set a route start to W Rabdolph 100 in Chicago. Simulate some typo corrections
await box.feed(text="chcago\b\b\b\bicago", delay=0.7)
box.submit()
await asyncio.sleep(2)
```

```python
await box.feed(text="w randolph 100", delay=0.08)
box.submit()
await asyncio.sleep(3)
```

```python
route_ctl.set_route_start(tuple(response_items[0]["access"][0].values()))
route_ctl.set_route_at(tuple(response_items[0]["access"][0].values()))
route_ctl.set_route_width(400)
```

```python
# Set a route end to W Roosevelt 3321.
await box.feed(text="w roosevelt 3321", delay=0.08)
box.submit()
await asyncio.sleep(3)
route_ctl.set_route_stop(tuple(response_items[0]["access"][0].values()))
await asyncio.sleep(2)
```

```python
# Search for gas stations on that route
await box.feed(text="gas stations", delay=0.08)
box.submit()
```
